```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/cbtn-data-science-ml/python-pandas-for-data-manipulation/main/starter_files/datasets/StudentsPerformance.csv")
df.head(3)
```

```python
df.info()
```

```python
df['gender'] = df['gender'].astype('category')
df['race/ethnicity'] = df['race/ethnicity'].astype('category')
df['parental level of education'] = df['parental level of education'].astype('category')
df.info()
```

```python
df['math score'] = pd.to_numeric(df['math score'].str.replace('%', ''), errors='coerce', downcast='float')
df['reading score'] = pd.to_numeric(df['reading score'].str.replace('%', ''), errors='coerce', downcast='float')
df['writing score'] = pd.to_numeric(df['writing score'].str.replace('%', ''), errors='coerce', downcast='float')
```
