# Searching and extracting: `find_all()` PART III

```python
# import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Get HTML into Python

```python
# assign HTML url to varible: url
# url ="https://webscraper.io/test-sites/e-commerce/allinone/phones"
url ="https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"

# get the HTML from url and assign to variable: page
page = requests.get(url)
page # The HTTP 200 OK success status response code = request has succeeded.
```

```python
# grab the HTML as text(string)
# parses it into Python and assign to variable: soup
soup = BeautifulSoup(page.text,"lxml")
soup # retruns the HTML 
```

## Create a table with multiple column representing various tag values

```python
# get product names
product_name = soup.find_all('a', class_ = 'title')
product_name
```

```python
# get product prices
product_price = soup.find_all('h4', class_ = 'pull-right price')
product_price
```

```python
import re
# get product reviews
product_review = soup.find_all('p', class_ = re.compile('pull'))
product_review
```

```python
# get product description
product_description = soup.find_all('p', class_ = re.compile('description'))
product_description
```

## Create a `for` loops to gather everything

```python
product_name_list = []

for i in product_name:
  name = i.text
  product_name_list.append(name)


product_name_list
```

```python
product_price_list = []

for i in product_price:
  price = i.text
  product_price_list.append(price)

product_price_list
```

```python
product_review_list = []

for i in product_review:
  review = i.text
  product_review_list.append(review)

product_review_list
```

```python
product_description_list = []

for i in product_description:
  description = i.text
  product_description_list.append(description)

product_description_list
```

## Convert to pandas DataFrame

```python
import pandas as pd
```

```python
table = pd.DataFrame({'Product Name':product_name_list, 'Product Price':product_price_list, 'Product Review':product_review_list, 'Product Description':product_description_list})
```
